## Introduction

This notebook depicts how images can be processed using openCV

## Setup

```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

## Reading image with cv2

```python
input_image = cv2.imread("DATA/00-puppy.jpg")
plt.imshow(input_image)
```

As we can see this image is nor RGB this image is actually BGR this can be converted to RGB using open CV only.

## Converting to RGB

```python
fix_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)
plt.imshow(fix_image)
```

## Convert to Grayscale

```python
image_gray = cv2.imread('DATA/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(image_gray, cmap='gray')
```

## Resizing

```python
plt.imshow(cv2.resize(image_gray, (500, 600)), cmap='gray')
```
