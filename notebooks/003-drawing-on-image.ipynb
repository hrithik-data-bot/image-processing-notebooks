## Introduction

This notebook shows how to draw on images

## Setup

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

## Blank Image

```python
blank_image = np.zeros(shape=(512, 512, 3), dtype=np.int16)
blank_image.shape
```

```python
plt.imshow(blank_image)
```

## Drawing rectange on image

```python
cv2.rectangle(img=blank_image,
              pt1=(100,100), # Top Left
              pt2=(400,200), # Bottom Right
              color=(255,0,0),
              thickness=5)
```

```python
plt.imshow(blank_image)
```

## Drawing square on an image

```python
cv2.rectangle(blank_image, pt1=(100, 300), pt2=(200, 400), color=(10,40,32), thickness=10)
```

```python
plt.imshow(blank_image)
```

## Circles

```python
cv2.circle(blank_image, center=(400,400), radius=50, color=(10,120,200), thickness=4)
plt.imshow(blank_image)
```

```python
cv2.circle(blank_image, center=(300, 350), radius=50, color=(68,45,12), thickness=-1)
plt.imshow(blank_image)
```

## Line

```python
cv2.line(img=blank_image, pt1=(80,100), pt2=(512, 0), color=(102,255,255), thickness=2)
plt.imshow(blank_image)
```
