Image Blending 

## Setup

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

## Images

```python
img_1 = cv2.cvtColor(cv2.imread("DATA/dog_backpack.png"), cv2.COLOR_BGR2RGB)
img_2 = cv2.cvtColor(cv2.imread("DATA/watermark_no_copy.png"), cv2.COLOR_BGR2RGB)
```

```python
plt.subplot(1,2,1)
plt.imshow(img_1)
plt.title('Dog Backpack Image')

plt.subplot(1,2,2)
plt.imshow(img_2)
plt.title('Do Not Copy Image')

plt.suptitle("Image 1 and Image 2")

plt.tight_layout()
```

## Resizing Images

```python
img_1 = cv2.resize(img_1, (1200, 1200))
img_2 = cv2.resize(img_2, (1200, 1200))
```

```python
plt.subplot(1,2,1)
plt.imshow(img_1)
plt.title('Dog Backpack Image Resized')

plt.subplot(1,2,2)
plt.imshow(img_2)
plt.title('Do Not Copy Image Resized')

plt.suptitle("Image 1 and Image 2 Resized")

plt.tight_layout()
```

## Blending using addWeight

```python
blended = cv2.addWeighted(img_1, 0.7, img_2, 0.3, 0)
plt.imshow(blended)
```

## Overlay Images

```python
img_2 = cv2.resize(img_2, (600, 600))
```

```python
plt.subplot(1,2,1)
plt.imshow(img_1)
plt.title('Dog Backpack Image')

plt.subplot(1,2,2)
plt.imshow(img_2)
plt.title('Do Not Copy Image Resized')

plt.suptitle("Image 1 and Image 2 Resized")

plt.tight_layout()
```

```python
small_image, large_image = img_2, img_1
```

```python
y_offset, x_offset = 0, 0
y_end, x_end = y_offset + small_image.shape[0], x_offset + small_image.shape[1]
large_image[y_offset:y_end, x_offset:x_end] = small_image
```

```python
plt.imshow(large_image)
```
