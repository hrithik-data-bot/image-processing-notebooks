## Introduction

This notebook depicts text and polygons on the image

## Setup

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

## Image

```python
image = np.zeros((500,1000, 3), np.int16)
plt.imshow(image)
```

## Text

```python
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(image, "Hrithik Sharma", (200, 400), font, 2, (200,34,123), 5, cv2.LINE_AA)
plt.imshow(image)
```

## Polygon

```python
vertices = np.array([[200,300], [400, 100], [600, 300]], np.int32).reshape((-1,1,2))
cv2.polylines(image, [vertices], isClosed=True, color=(255, 0, 0), thickness=5)
plt.imshow(image)
```
